```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Formating dataset

```python
col_name = ['Device ID', 'Time', 'Temperature', 'Humidity', 'CO2']
df = pd.read_table("intelli_gee_sensor_THC_10K.dat", sep=',', names=col_name)
```

```python
df.head()
```

```python
for col in col_name:
    df[col] = df[col].apply(lambda x: x.split()[1])
```

```python
df.head()
```

```python
# To get rid of the bracket at the end
df['CO2'] = df['CO2'].apply(lambda x: x[:-1]) 

# Converting columns with numerical string into float
for col in col_name[1:]:
    df[col] = df[col].apply(lambda x: float(x))
```

```python
df.head()
```

## Overview and Visualization of Data in Time Series

```python
df.info()
```

```python
df.describe()
```

```python
# Correlation matrix
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
df.plot(x=1,y=4,figsize=(18,10),title='Time Series of CO2')
plt.show()
```

```python
# Calculating Moving Average
df['Temperature_ma'] = df['Temperature'].rolling(1000).mean()
df['Humidity_ma'] = df['Humidity'].rolling(1000).mean()
df['CO2_ma'] = df['CO2'].rolling(1000).mean()
```

```python
df.tail()
```

```python
# New plots with Moving Average set to window=1000

df.plot(x=1, y=5,figsize=(18,10), title='Temperature with MA')
df.plot(x=1, y=6,figsize=(18,10), title='Humidity with MA')
df.plot(x=1, y=7,figsize=(18,10), title='CO2 with MA')

plt.show()
```
